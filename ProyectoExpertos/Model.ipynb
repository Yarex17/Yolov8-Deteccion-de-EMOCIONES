```python
import cv2
from ultralytics import YOLO

# Diccionario para mapear los nombres de las clases en inglÃ©s a espaÃ±ol
class_names_es = {
    'angry': 'enojado',
    'disgust': 'disgusto',
    'fear': 'miedo',
    'happy': 'feliz',
    'neutral': 'neutral',
    'sad': 'triste',
    'surprised': 'sorprendido'
}

# Cargar el modelo entrenado
model = YOLO(model_path)

# Inicializar la cÃ¡mara
cap = cv2.VideoCapture(0)  # 0 para la cÃ¡mara predeterminada

while True:
    # Capturar una imagen de la cÃ¡mara
    ret, frame = cap.read()

    # Realizar detecciÃ³n de emociones en la imagen
    results = model(frame)

    # Procesar los resultados
    for result in results:
        boxes = result.boxes
        for box in boxes:
            # Obtener la clase (emociÃ³n) y la probabilidad
            class_id = int(box.cls)
            class_name_en = model.names[class_id]
            class_name_es = class_names_es[class_name_en]  # Traducir el nombre de la clase a espaÃ±ol
            probability = float(box.conf)

            # Dibujar el rectÃ¡ngulo delimitador y el nombre de la clase en la imagen
            x1, y1, x2, y2 = box.xyxy.tolist()[0]
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            cv2.putText(frame, f"{class_name_es} ({probability:.2f})", (int(x1), int(y1 - 10)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (36, 255, 12), 2)

    # Mostrar la imagen con las detecciones
    cv2.imshow('Deteccion de Emociones', frame)

    # Salir del bucle si se presiona la tecla 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberar recursos
cap.release()
cv2.destroyAllWindows()
```
